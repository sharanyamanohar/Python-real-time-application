difference between loc & iloc
loc-
1) loc is label based indexing operations,selecting a data in basis of row and column name
2) You can use slices, lists, or boolean arrays to select specific rows and columns.
3) But it takes two arguements ;the row label and column label
4) Imp-The row and column label is inclusive i.e both the start and end label that we are giving it      includes in selection operation.
5) The syntax will be of (loc) ** (df.loc[row,column]).

iloc-
1) iloc is integer based indexing operation,select data from a DataFrame based on integer positions      (row and column indices)
2) You can use slices or lists of integers to select specific rows and columns.
3) But it takes two arguements ;the row label and column label
4) Imp-The row and column indices are exclusive for the end value i.e, the end value is not included      in the selection
5) The syntax will be of (iloc) ** (df.iloc[row,column]).
